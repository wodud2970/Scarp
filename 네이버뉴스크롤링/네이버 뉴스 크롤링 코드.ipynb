```python
from bs4 import BeautifulSoup     
from selenium import webdriver
import time
import pandas as pd

# 크롤링 할 url_list 추출하기

path = "./chromedriver.exe"
driver = webdriver.Chrome(path)

dt_index = pd.date_range(start='20210625', end='20210725')
dt_list = dt_index.strftime("%Y.%m.%d").tolist()
query_txt = "부산"
page = 10 # 크롤링할 페이지 수

url_list = [] # url주소

for date in dt_list:
    url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query={0}&sort=0&photo=0&field=0&pd=3&ds={1}&de={1}&cluster_rank=29&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:from20210723to20210723,a:all&start=1".format(query_txt, date)
    driver.get(url)
    time.sleep(2)

    for _ in range(page): 
        html = driver.page_source # 현재화면의 html을 모두 불러옴
        soup = BeautifulSoup(html, 'html.parser') # 파싱 작업
        ul = soup.select("#main_pack > section.sc_new.sp_nnews._prs_nws > div > div.group_news > ul")
        li = ul[0].find_all("li",{"class":"bx"})

        for i in li:
            # 네이버 뉴스만 가져오기
            try:
                href = i.find_all("a",{"class":"info"})[1].get("href")
                if "https://sports.news.naver.com/" not in href: # 스포츠 뉴스 거르기
                    url_list.append(href)
            except:
                continue

        driver.find_element_by_xpath('''//*[@id="main_pack"]/div[2]/div/a[2]/i''').click()
   
```

```python
for idx, url in enumerate(url_list):
    print("index = ",idx,"url = ", url)
```

```python
# 뉴스 기사 제목, 본문 긁기

title_list = [] # 뉴스기사제목
content_list = [] # 뉴스기사내용

for idx, url in enumerate(url_list):
    driver.get(url)
    time.sleep(1)
    
    html = driver.page_source # 현재화면의 html을 모두 불러옴
    soup = BeautifulSoup(html, 'html.parser') # 파싱 작업
    
    try:
        # 제목
        title = soup.find("div", {"class" : "article_info"}).find("h3", {"id" : "articleTitle"}).get_text()
        # 본문
        content = soup.find("div",{"id" : "articleBodyContents"}).get_text().replace("// flash 오류를 우회하기 위한 함수 추가","").replace("function _flash_removeCallback() {}","").strip()
        # 본문 요약 정보 제거
        summary = soup.find("div",{"id" : "articleBodyContents"}).find("strong", {"class" : "media_end_summary"}) 
        if summary == None:
            pass
        else:
            content = content.replace(str(summary.get_text()),"").strip()

        title_list.append(title)
        content_list.append(content)


        print("{}번째 완료".format(idx))
    except:
        print("{}번째 실패".format(idx))
```

```python
# 저장하기
df = pd.DataFrame({"title":title_list, "content":content_list})

df.to_csv('./네이버뉴스크롤링.txt', sep = '\t', index = False)
```

```python
# MeCab을 window에서 쓰기위한 함수
"""
    Parsing 규칙의 문제점, split을 "," 기준으로 하는데, token이 "," 인 경우에는 쉼표만 잘려서 나오기 때문에, 
    + "%," 같이 특수문자와 쉼표가 같이 등장하는 경우도 생각해주어야 함.
    
    (",", "SC") 의 원래 튜플이 만들어지지 않음.
    
    명사 분석의 경우 해당 토큰이 필요하지 않으니 pass
    
    형태소 분석과 POS tagging의 경우 해당 토큰이 필요하므로, token[0]이 ' 인 경우엔 따로 (",", "SC")를 집어 넣어줘야함.
"""
import MeCab
mecab = MeCab.Tagger()

def mecab_nouns(text): # 명사만 추출
    nouns = []
    
    # 우리가 원하는 TOKEN\tPOS의 형태를 추출하는 정규표현식.
    pattern = re.compile(".*\t[A-Z]+") 
    
    # 패턴에 맞는 문자열을 추출하여 konlpy의 mecab 결과와 같아지도록 수정.
    temp = [tuple(pattern.match(token).group(0).split("\t")) for token in mecab.parse(text).splitlines()[:-1]]
        
    # 추출한 token중에 POS가 명사 분류에 속하는 토큰만 선택.
    for token in temp:
        if token[1] == "NNG" or token[1] == "NNP" or token[1] == "NNB" or token[1] == "NNBC" or token[1] == "NP" or token[1] == "NR":
            nouns.append(token[0])
        
    return nouns

def mecab_morphs(text): # 형태소만 추출
    morphs = []
    
    # 우리가 원하는 TOKEN\tPOS의 형태를 추출하는 정규표현식.
    pattern = re.compile(".*\t[A-Z]+") 
    
    # 패턴에 맞는 문자열을 추출하여 konlpy의 mecab 결과와 같아지도록 수정.
    temp = [tuple(pattern.match(token).group(0).split("\t")) for token in mecab.parse(text).splitlines()[:-1]]
        
    # 추출한 token중에 문자열만 선택.
    for token in temp:
        morphs.append(token[0])
    
    return morphs

def mecab_pos(text): # 형태소, 품사 추출
    pos = []
    # 우리가 원하는 TOKEN\tPOS의 형태를 추출하는 정규표현식.
    pattern = re.compile(".*\t[A-Z]+") # 한번쓸게 아니라 여러번쓸것이기 때문에 컴파일하면 속도면에서 빠르다.
    
    # 패턴에 맞는 문자열을 추출하여 konlpy의 mecab 결과와 같아지도록 수정.
    pos = [tuple(pattern.match(token).group().split("\t")) for token in mecab.parse(text).splitlines()[:-1]] # https://dojang.io/mod/page/view.php?id=2437 참고
        
    return pos
```

```python
def text_cleaning(doc):
    # 한국어를 제외한 글자를 제거하는 함수.
    doc = re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]", "", doc)
    return doc

def define_stopwords(path):
    
    SW = set()
    # 불용어를 추가하는 방법 1.
    # SW.add("있다")
    
    # 불용어를 추가하는 방법 2.
    # stopwords-ko.txt에 직접 추가
    
    with open(path, encoding='utf-8') as f:
        for word in f:
            SW.add(word)
            
    return SW

def text_tokenizing(doc):
#     return [word for word in mecab_morphs(doc) if word not in SW and len(word) > 1]
    
    # wordcloud를 위해 명사만 추출하는 경우.
    return [word for word in mecab_nouns(doc) if word not in SW and len(word) > 1]
```

```python
import json
import os
import re
from pprint import pprint

news_list = []

for i in df["tc"]:
    text_cleaning(i)
    news_list.append(i)

# 불용어 처리
SW = define_stopwords("./stopwords-ko.txt")

clean_news_list = []
for news in news_list:
    for i in SW:
        news.replace(i, "")
    clean_news_list.append(news)
    

# 시간이 오래 걸려서 한번 했으면 그파일을 저장하고 다음에 할때 파일을 여는 것으로 하기 위해서 이렇게함. 

if os.path.exists('news_data.json'): # 이 파일이 존재하면 True 없으며 False
    with open("news_data.json", encoding='utf-8') as f:
        news_data = json.load(f)
else:
    news_data = [(text_tokenizing(line)) for line in clean_news_list]
    
    
    with open("news_data.json", 'w', encoding='utf-8') as f:
        json.dump(news_data, f, ensure_ascii=False, indent='\t') # indent='\t' 탭 기준으로 저장한다.

        
# pretty print 프린트가 조금더 깔끔하게 나오는것
pprint(news_data)
```

```python
news_data2 = []
for i in news_data:
    for j in i:
        news_data2.append(j)
```
