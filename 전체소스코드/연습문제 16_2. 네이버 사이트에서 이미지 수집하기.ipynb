```python
#연습문제 : 네이버 이미지 수집용 크롤러

#Step 1. 필요한 모듈과 라이브러리를 로딩합니다.

from bs4 import BeautifulSoup
from selenium import webdriver
import urllib.request
import urllib
import time
import sys
import re
import math
import time
import os
import random

#Step 2. 사용자에게 필요한 정보들을를 입력 받습니다.

print("=" *80)
print(" 이 크롤러는 네이버 이미지 정보를 수집합니다")
print("=" *80)
print("\n")

query_txt = input('1.크롤링할 이미지의 키워드는 무엇입니까?: ')
start_date = input('2.조회 시작일자 입력(예:2017-01-01):')
end_date = input('3.조회 종료일자 입력(예:2017-12-31):')

cnt = int(input('4.크롤링 할 건수는 몇건입니까?: '))
real_cnt = math.ceil(cnt / 50)

f_dir=input('5.파일이 저장될 경로만 쓰세요(예: c:\\temp\\ ) : ')
if f_dir =='' :
        f_dir = "c:\\temp\\"

now = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

os.chdir(f_dir)
os.makedirs(f_dir+s+'-'+query_txt)
os.chdir(f_dir+s+'-'+query_txt)
f_result_dir = f_dir+s+'-'+query_txt
        
#Step 3. 크롬 드라이버를 사용해서 웹 브라우저를 실행합니다.

s_time = time.time( )

#Step 3. 크롬 드라이버를 사용해서 웹 브라우저를 실행합니다.
path = "c:/temp/chromedriver_240/chromedriver.exe"
driver = webdriver.Chrome(path)


driver.get('https://www.naver.com')
time.sleep(random.randrange(2,5))

element = driver.find_element_by_id("query")

element.send_keys(query_txt)
element.submit()

#Step 4. 아래의 블로그 링크를 선택합니다

driver.find_element_by_link_text("이미지").click()

#Step 5 검색 옵션 버튼 클릭
driver.find_element_by_xpath("""//*[@id="_search_option_btn"]""").click()


#Step 7. 기간 버튼 클릭
driver.find_element_by_xpath("""//*[@id="snb"]/div/ul/li[1]/a""").click()
                              
time.sleep(2)

# 직접 입력 시작 날짜 클릭
s_date = driver.find_element_by_xpath("""//*[@id="img_input_startdate"]""")
                                         
driver.find_element_by_xpath("""//*[@id="img_input_startdate"]""").click()
s_date.clear()
time.sleep(1)

for c in start_date:
    s_date.send_keys(c)
    time.sleep(0.3)
                                          
e_date = driver.find_element_by_xpath("""//*[@id="img_input_enddate"]""")
driver.find_element_by_xpath("""//*[@id="img_input_enddate"]""").click()
e_date.clear()
time.sleep(1)

for c in end_date:
    e_date.send_keys(c)
    time.sleep(0.3)

driver.find_element_by_xpath("""//*[@id="_nx_option_date"]/div/span/button/span""").click()
                                

time.sleep(2)


# 스크롤 다운 함수 만들기

def scroll_down(driver):
  driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
  time.sleep(3)


i = 1
while (i <= real_cnt):
      scroll_down(driver) 
      i += 1

      if i ==  6 :
         driver.find_element_by_xpath("""//*[@id="_sau_imageTab"]/div[2]/div[8]/a""").click()


# Step 8. 이미지 추출하기  

file_no = 0
count = 1
img_src2=[]



html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')


imgs = driver.find_elements_by_tag_name('img')

for img in imgs:
        img_src1=img.get_attribute('src')
        img_src2.append(img_src1)
        count += 1
        

        if count == cnt+10 :
            break

for i in range(3,len(img_src2)+1) :

        file_no += 1
        urllib.request.urlretrieve(img_src2[i],str(file_no)+'.jpg')       
        time.sleep(1)
        print("\n")
        print("%s 번째 이미지 저장중입니다=======" %file_no)
        print("\n")

        if file_no == cnt :
                break

        
e_time = time.time( )
t_time = e_time - s_time

store_cnt = file_no -1

print("=" *70)
print("총 소요시간은 %s 초 입니다 " %round(t_time,1))
print("총 저장 건수는 %s 건 입니다 " %file_no)
print("파일 저장 경로: %s 입니다" %f_result_dir)
print("=" *70)


#driver.close( )
```
