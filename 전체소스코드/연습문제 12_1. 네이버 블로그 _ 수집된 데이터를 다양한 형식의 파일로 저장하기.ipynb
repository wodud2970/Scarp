```python
# 3) 저장할 내용을 목록으로 만들어서  엑셀형식과 csv 형식으로 저장하기

print("=" *80)
print(" 연습문제 : 저장할 내용을 목록으로 만들어서 xls , csv 형식으로 저장하기")
print("=" *80)

#Step 0. 필요한 모듈과 라이브러리를 로딩하고 검색어를 입력 받습니다

from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys        # system 설정을 변경하기 위해 필요합니다


import pandas  as pd    

query_txt = input('1.크롤링할 키워드는 무엇입니까?: ')

f_name=input('2.txt 형태로 저장할 경로와 파일명을 확장자 포함해서 쓰세요:')
fc_name=input('3.csv 형태로 저장할 경로와 파일명을 확장자 포함해서 쓰세요: ')
fx_name=input('4.xls 형태로 저장할 경로와 파일명을 확장자 포함해서 쓰세요: ')


#Step 1. 크롬 드라이버를 사용해서 웹 브라우저를 실행합니다.
path = "c:/temp/chromedriver_240/chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get('http://www.naver.com')
time.sleep(2)

#Step 2. 네이버 검색창에 입력 받은 검색어를 넣고 검색한 후 "블로그" 선택
element = driver.find_element_by_id("query")
element.send_keys(query_txt)
element.submit()

driver.find_element_by_link_text("블로그").click( )

# Step 9. 저장 목록을 만든 후 목록에 있는 내용을 파일에 저장하기

no2 = [ ]           # 게시글 번호 컬럼
title2 = [ ]        # 게시물 제목 컬럼
contents2 = [ ]     # 게시글 내용 컬럼
bdate2 = [ ]        # 작성 일자 컬럼
nick2 = [ ]         # 블로그 닉네임

no = 1

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
blog_list = soup.select('ul > li > dl')

for i in blog_list:
    f = open(f_name, 'a',encoding='UTF-8')
    
    print(no)                                 # 게시물 번호
    no2.append(no)                            # 게시물 번호 리스트에 추가
    f.write('1.번호:' + str(no) + "\n" ) 
    
    title = i.find('dt').get_text( )          # 게시물 제목
    title2.append(title)                      # 게시물 제목 리스트에 추가
    f.write('2.제목:' + str(title)+ "\n" ) 
    print(title)
   
    contents = i.find('dd' , 'sh_blog_passage').get_text( )   # 게시물 내용
    contents2.append(contents)                # 게시물 내용 리스트에 추가
    f.write('3.내용:' + str(contents) + "\n" )
    print(contents)

    bdate = i.find('dd','txt_inline').get_text( )  # 작성일자
    bdate2.append(bdate)                     # 작성일자 리스트에 추가
    f.write('4.작성일자:' + str(bdate) + "\n" )
    print(bdate)
    
    nick = i.find('dd','txt_block').find('a').get_text( )
    nick2.append(nick)
    f.write('5.블로그닉네임:' + str(nick) + "\n" )
    print(nick)
    
    f.write("=" * 80 + "\n")
    f.close( )
    
    print("\n")
    
    no += 1
    
    

# 출력 결과를 표(데이터 프레임) 형태로 만들기

naver_blog = pd.DataFrame()
naver_blog['번호'] = no2
naver_blog['제목'] = title2
naver_blog['내용'] = contents2
naver_blog['작성일자'] = bdate2
naver_blog['블로그닉네임'] = nick2

# csv 형태로 저장하기
naver_blog.to_csv(fc_name,encoding="utf-8-sig",index=False)
print(" csv 파일 저장 경로: %s" %fc_name) 

# 엑셀 형태로 저장하기
naver_blog.to_excel(fx_name , index=False)
print(" xls 파일 저장 경로: %s" %fx_name) 

# 출력 결과를 txt 파일로 저장하기
print(" txt 파일 저장 경로: %s" %f_name)  

```
