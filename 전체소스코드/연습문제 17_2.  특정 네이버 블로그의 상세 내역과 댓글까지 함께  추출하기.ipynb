```python
# 연습문제 : 네이버 블로그 정보 추출하여 다양한 형식의 파일로 저장하기

print("=" *80)
print(" 연습문제 : 네이버 블로그 상세 내역과 댓글 정보 추출하여 저장하기")
print("=" *80)


#Step 1. 필요한 모듈과 라이브러리를 로딩합니다.
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys
import re
import math
import numpy  # pip install numpy 를 해야 사용가능함
import pandas as pd  # pip install pandas 해야 사용가능하고 numpy 가 먼저 설치되어 있어야

import random
import os   # 폴더 생성하기 위해 필요함

#Step 2. 사용자로부터 필요한 정보 입력받기
blog_url = input('1.크롤링할 블로그 주소를 입력하세요: ')
query_txt1 = '블로그댓글수집'
    
f_dir = input("2.결과 파일을 저장할 폴더명만 쓰세요(예:c:\\temp\\):")
print("\n")

print("요청하신 데이터를 수집 중입니다")
print("\n")
print("잠시만 기다려 주세요~~~~~^^")
print("\n")
    
# 저장될 파일위치와 이름을 지정합니다
now = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

os.makedirs(f_dir+s+'-'+query_txt1)
os.chdir(f_dir+s+'-'+query_txt1)

ff_name=f_dir+s+'-'+query_txt1+'\\'+s+'-'+query_txt1+'.txt'
fc_name=f_dir+s+'-'+query_txt1+'\\'+s+'-'+query_txt1+'.csv'
fx_name=f_dir+s+'-'+query_txt1+'\\'+s+'-'+query_txt1+'.xls'

#Step 3. 크롬 드라이버를 사용해서 웹 브라우저를 실행합니다.

s_time = time.time( )

path = "c:/temp/chromedriver_240/chromedriver.exe"
driver = webdriver.Chrome(path)


blog_addr2=[]
w_name2=[]
w_date2=[]
blog_txt2=[]

reple_1st_name=[]    # 1차 댓글 작성자이름
reple_1st_txt=[]     # 1차 댓글 내용
reple_1st_date=[]    # 1차 댓글 날짜

reple_2nd_name=[]    # 2차 댓글 작성자이름
reple_2nd_txt=[]     # 2차 댓글 내용
reple_2nd_date=[]    # 2차 댓글 날짜


gubun = re.split("[/]",blog_url)

if gubun[2] != 'blog.naver.com' :
    print(" 네이버 블로그만 가능합니다")
    
else :
    driver.get(blog_url)
    time.sleep(random.randrange(2,5))  # 2 - 5 초 사이에 랜덤으로 시간 선택
      
    #Step 9. 각 블로그의 상세 결과를 출력하여 파일에 저장하기

    print("\n")
    print("텍스트 추출 시작 =================================================")
    print("\n")

    #try: 
    driver.switch_to.frame('mainFrame')
    #except:
    #    continue
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    addr_1 = soup.select('#postViewArea')
    addr_2 = soup.select('div[class="se-main-container"]')
       

    if addr_1 :

            f = open(ff_name, 'a',encoding='UTF-8')

            print("=" *80)             
            print("1.블로그주소: ",blog_url)
            blog_addr2.append(blog_url)
            f.write("\n")

            f.write("블로그 데이터를 수집합니다==============" + "\n")
           
            f.write("1.블로그 주소:"+blog_url + "\n")

            writer = soup.select("div.blog2_container > span.writer")
            try :
                wname = writer[0].get_text( )   # 작성자 닉네임
            except IndexError :
                wname = " "
            else :
                wname = wname.replace("\n","")
            print("2.작성자 닉네임: ",wname )
            f.write("2.작성자 닉네임:" + wname + "\n")
            w_name2.append(wname)

            wdate = soup.select('p[class="date fil5 pcol2 _postAddDate"]')
            try : 
                wdate = wdate[0].get_text( )
            except IndexError :
                wdate = ' '
            print("3.작성일자:",wdate)
            f.write("3.작성 일자:" + wdate + "\n")
            w_date2.append(wdate)

            for i in addr_1:
                for tag in soup.find_all('div',{'class':'se_component se_map default'}) :
                #for tag in soup.find_all('div',{'class':'se_component se_oglink og_bSize '}) :
                            tag.decompose()
                for tag2 in soup.find_all('div',{'class':'se_component se_oglink default '}) :
                            tag2.decompose()
                blog_txt = i.text.replace("\n","")

                try :
                    print("4.블로그내용: \n",blog_txt)
                except UnicodeEncodeError :
                    continue
                    
                if len(blog_txt) > 32767 :
                    blog_txt = blog_txt[1:32760]
                    
                print("\n")
                f.write("4.블로그 내용:" + blog_txt + "\n")
                blog_txt2.append(blog_txt)
                
                
            # 댓글 정보 추출하기
  
                    
            time.sleep(1)
                
                
            f.close( )
                

    elif addr_2 :
            
            print(" 블로그 데이터를 수집합니다==============" )
            
            f = open(ff_name, 'a',encoding='UTF-8')
            print("=" *80)    
            print("1.블로그주소: ",blog_url)
            blog_addr2.append(blog_url)
            f.write("\n")
            f.write("블로그 데이터를 수집합니다=============="  + "\n")
            
            f.write("1.블로그 주소:"+blog_url + "\n")

            writer = soup.select("div.blog2_container > span.writer")
            try :
                wname = writer[0].get_text( )   # 작성자 닉네임
            except IndexError :
                wname = " "
            else :
                wname = wname.replace("\n","")
                    
            print("2.작성자 닉네임: ",wname )
            f.write("2.작성자 닉네임:" + wname + "\n")
            w_name2.append(wname)

            wdate = soup.select("div.blog2_container > span.se_publishDate.pcol2")
            try : 
                wdate = wdate[0].get_text( )
            except IndexError :
                wdate = " "
            
            print("3.작성일자:",wdate)
            f.write("3.작성 일자:" + wdate + "\n")
            w_date2.append(wdate)
                                  
            for i in addr_2:

                map_1 = soup.select('div[class="se_component se_map default"]')
                map_2 = soup.select('div[class="se_component se_oglink og_bSize "]')

                if map_1 :
                    for tag in soup.find_all('div',{'class':'se_component se_map default'}) :
                            tag.decompose()

                elif map_2 :
                    for tag2 in soup.find_all('div',{'class':'se_component se_oglink og_bSize '}) :
                            tag2.decompose()               

                blog_txt = i.text.replace("\n","")
                
                # 블로그 내용 출력중에 UnicodeEncodeError: 'UCS-2' codec can't encode characters.. 라는
                # 에러가 나올 경우가 있다. 이럴 때를 대비해서 예외처리를 해야 한다
                try :
                    print("4.블로그내용: \n",blog_txt)
                except UnicodeEncodeError :
                    continue
                    
                if len(blog_txt) > 32767 :
                    blog_txt = blog_txt[1:32760]

                print("\n")
                f.write("4.블로그 내용:" + blog_txt + "\n")
                blog_txt2.append(blog_txt)
                
                # 댓글 정보 추출하기
                try :
                    driver.find_element_by_xpath('''//*[@id="Comi221514204265"]/span/i''').click( )    
                    #driver.find_element_by_class_name('bu_arr').click( )                                                 
                except :
                    print('이 글의 댓글이 없습니다')    
                    print("\n")
                    continue                    
                else :
                    time.sleep(1)
            
                    html2 = driver.page_source
                    soup2 = BeautifulSoup(html2, 'html.parser') 
            
                    reple_1 = soup2.select('div.u_cbox_content_wrap > ul.u_cbox_list > li ') 
        
                    for a in range(0,len(reple_1)):
                        print(" ==게시글에 작성된 1차 댓글을 출력합니다. ==========================")
                        reple_name = reple_1[a].find('span','u_cbox_nick').get_text()
                        print('1.댓글작성자: ',reple_name)
                        f.write("5.댓글작성자:" + reple_name + "\n")
                        reple_1st_name.append(reple_name)
                                              
                        reple_txt = reple_1[a].find('div','u_cbox_text_wrap').get_text()  
                        print('2.댓글내용: ',reple_txt)
                        f.write("6.댓글내용:" + reple_txt + "\n")
                        reple_1st_txt.append(reple_txt)
                        
                        reple_date = reple_1[a].find('div','u_cbox_info_base').find('span','u_cbox_date').get_text()
                        print('3.댓글작성일자: ',reple_date)
                        f.write("7.댓글작성일자:" + reple_date + "\n")
                        reple_1st_date.append(reple_date)
                        
                        try :
                            reple_2 = reple_1[a].select('div.u_cbox_reply_area > ul > li')
                        except :
                            print("\n")
                            reple2_txt='비밀글입니다'
                            f.write("=====8.답글내용:" + reple2_txt + "\n")
                            print("\n")
                            reple_2nd_txt.append(reple2_txt)
                        else :
            
                            for b in range(0,len(reple_2)):   
                                try :
                                    reple2_name = reple_2[b].find('span','u_cbox_nick').get_text()
                                except :
                                    print("\n")
                                    reple2_name='비밀글입니다'
                                    f.write("=====9.답글작성자:" + reple2_name + "\n")
                                    print(reple2_name)
                                    print("\n")
                                    reple_2nd_name.append(reple2_name)
                                else :
                                    print('=====1.답글작성자:',reple2_name)
                                    f.write("=====9.답글작성자:" + reple2_name + "\n")
                                    reple_2nd_name.append(reple2_name)
                                    
                                    reple2_txt = reple_2[b].find('span','u_cbox_contents').get_text()  
                                    print('=====2.답글내용:',reple2_txt)
                                    f.write("=====10.답글내용:" + reple2_txt + "\n")
                                    reple_2nd_txt.append(reple2_txt)
                                    
                                    reple2_date = reple_2[b].find('span','u_cbox_date').get_text()
                                    print('=====3.답글작성일자:',reple2_date)
                                    f.write("=====11.답글작성자:" + reple2_date + "\n")
                                    reple_2nd_date.append(reple2_date)
                
                
            f.close()

            
```

```python
e_time = time.time( )
t_time = e_time - s_time


print("\n") 
print("=" *100)
print("총 소요시간은 %s 초 입니다 " %round(t_time,1))
print("파일 저장 완료: txt 파일명 : %s " %ff_name)
print("파일 저장 완료: csv 파일명 : %s " %fc_name)
print("파일 저장 완료: xls 파일명 : %s " %fx_name)
print("=" *100)

driver.close()
```
