```python
# 구글에서 pdf 파일 추출하여 다운로드 받기
#Step 1. 필요한 모듈과 라이브러리를 로딩합니다.

from bs4 import BeautifulSoup
from selenium import webdriver
import urllib.request
import urllib
import time
import sys
import re
import math
import os
import random



#Step 2. 사용자에게 검색어 키워드를 입력 받습니다.

print("=" *80)
print(" 연습문제 : 구글 사이트에서 pdf 파일을 검색하여 수집하는 크롤러 ")
print("=" *80)

query_txt = input('   1.크롤링할 키워드는 무엇입니까?: ')
cnt = int(input('   2.크롤링 할 건수는 몇건입니까?: '))

  
f_dir=input('   3.파일이 저장될 경로만 쓰세요(예: c:\\temp\\ ) : ')

#Step 3. 파일을 저장할 폴더를 생성합니다


if  os.path.isdir(f_dir) :
    print("\n")
    print('   입력하신 경로가 존재하여  %s 폴더에 저장하겠습니다' %f_dir)
else :       
    os.makedirs(f_dir)
    print("\n")
    print('   입력하신 경로가 존재하지 않아 %s 폴더를 생성했습니다' %f_dir)

print("\n")
print("   요청하신 데이터를 수집 중이오니 잠시만 기다려 주세요~~^^")

#Step 4. 크롬 드라이버를 사용해서 웹 브라우저를 실행한 후 검색합니다

s_time = time.time( )

options = webdriver.ChromeOptions()

path = "c:/temp/chromedriver_240/chromedriver.exe"

options.add_experimental_option("prefs", {

      "download.default_directory": f_dir,
      "download.prompt_for_download": False,
      "plugins.always_open_pdf_externally": True
})

driver = webdriver.Chrome(path,options=options)

driver.command_executor._commands["send_command"] = ("POST", '/session/$sessionId/chromium/send_command')
params = {'cmd': 'Page.setDownloadBehavior', 'params': {'behavior': 'allow', 'downloadPath': f_dir}}
command_result = driver.execute("send_command", params)

driver.get('https://www.google.com')
time.sleep(random.randrange(2,5))

element = driver.find_element_by_name("q")

element.send_keys(query_txt)
element.submit()


#Step 5. 각 게시글의 URL 주소를 추출하여 목록을 먼저 만든 후 상세 크롤링을 진행합니다.
# URL 수집하여 목록 만들기
click_count = 1
url_list=[]
count = 0

while ( count < cnt + 1) :     

        print("   %s 번째 페이지에서 pdf 파일을 수집합니다==============" %click_count )                 
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        result_1 = soup.select('div.bkWMgd > div.srg > div.g')
        
        # result_1 을 for 로 돌려서 pdf 포함된 글만 추려내서 url목록 만들기
        
        for i in result_1 :
            if (i.find('span').get_text( ) =='[PDF]') :
                url_list.append(i.find('a')['href'])
                driver.get(i.find('a')['href'])
                count += 1
                 
        print("  %s 번째 페이지까지 총 %s건 정보 수집 완료 ====================" %(click_count,count))
        print("\n")
        click_count += 1
        
        if count == cnt:
          break
        else :
            driver.find_element_by_link_text("%s" %click_count).click()
 
e_time = time.time( )
t_time = e_time - s_time

#Step 6. 요약 정보 보여주기
print("=" *100)
print("총 소요시간은 %s 초 입니다 " %round(t_time,1))
print("파일이 저장된 폴더명 : %s " %f_dir)
print("=" *100)
```
