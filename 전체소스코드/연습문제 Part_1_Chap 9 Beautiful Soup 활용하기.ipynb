```python
img_src_2017=[]
img_src_2018=[]

for c in img_src :
    img_src1=c['src']
    #print(img_src1)
    
    if '2017' in img_src1 :
        img_src_2017.append(img_src1)
            #print(img_src1)
    elif '2018' in img_src1 :
        img_src_2018.append(img_src1)
                
```
